```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style and figure size
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))

# Convert boolean values to strings for proper palette mapping
df['Revenue'] = df['Revenue'].astype(str)

# Total number of observations
total = float(len(df))

# Create a count plot with custom colors for True and False bars
ax = sns.countplot(x="Revenue", data=df, palette={'True': 'goldenrod', 'False': 'red'})

# Add data labels
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height() / total)
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the columns for grouping and hue
x, y = 'VisitorType', 'Weekend'

# Group data by 'VisitorType' and calculate percentage for 'Weekend'
df1 = df.groupby(x)[y].value_counts(normalize=True)
df1 = df1.mul(100)
df1 = df1.rename('percent').reset_index()

# Create a categorical plot with bars, specifying custom colors
palette = ['red', 'goldenrod']  # Define colors for the bars
g = sns.catplot(x=x, y='percent', hue=y, kind='bar', data=df1, palette=palette)

# Set the y-axis limit
g.ax.set_ylim(0, 100)

# Add data labels to each bar
for p in g.ax.patches:
    txt = f"{p.get_height():.2f}%"
    txt_x = p.get_x() + p.get_width() / 2
    txt_y = p.get_height()
    g.ax.text(txt_x, txt_y, txt, ha='center', va='bottom')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the columns for grouping and hue
x = 'TrafficType'
y = 'Revenue'

# Group data by 'TrafficType' and calculate percentage for 'Revenue'
df1 = df.groupby(x)[y].value_counts(normalize=True)
df1 = df1.mul(100)
df1 = df1.rename('percent').reset_index()

# Determine unique values in the 'Revenue' column
unique_values = df[y].unique()

# Define a palette for the bars based on unique values
palette = {value: 'red' if value == 'False' else 'gold' for value in unique_values}

# Create a categorical plot with bars and custom colors
g = sns.catplot(x=x, y='percent', hue=y, kind='bar', data=df1, palette=palette)

# Set the y-axis limit
g.ax.set_ylim(0, 100)

# Add data labels to each bar
for p in g.ax.patches:
    txt = f"{p.get_height():.2f}%"
    txt_x = p.get_x() + p.get_width() / 2
    txt_y = p.get_height()
    g.ax.text(txt_x, txt_y, txt, ha='center', va='bottom')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crosstab data
df = pd.crosstab(df['TrafficType'], df['Revenue'])

# Plotting the stacked bar chart with custom colors
df.div(df.sum(1).astype(float), axis=0).plot(kind='bar', stacked=True, color=['red', 'gold'])

# Adding title to the plot
plt.title('Traffic Type vs Revenue', fontsize=30)

# Display the plot
plt.show()
```
